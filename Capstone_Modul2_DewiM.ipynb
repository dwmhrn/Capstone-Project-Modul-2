## Outliers

Kita akan melakukan pengecekan Outliers pada kolom-kolom yang difokuskan pada penelitian kali ini:

```python
# Fungsi untuk mengidentifikasi outliers menggunakan metode IQR
def identify_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Kolom yang perlu diperiksa outliers
outlier_columns = ['Year_Birth', 'Income', 'MntWines', 'MntFruits', 
                   'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

# Mengidentifikasi outliers untuk setiap kolom yang relevan
outliers_dict = {col: identify_outliers(df, col) for col in outlier_columns}

# Menampilkan jumlah outliers untuk setiap kolom
outliers_count = {col: len(outliers_dict[col]) for col in outlier_columns}
outliers_count
```

## Pengeluaran Produk Berdasarkan Variabel Demografis

Sebelumnya, untuk mempermudah analisis, kita akan mengelompokkan pelanggan berdasarkan rentang usia mereka. Karena `Year_Birth` adalah variabel numerik, kita perlu mengonversinya menjadi kategori usia terlebih dahulu. 


1. **Penghitungan Usia:** 
   - Usia dihitung dengan mengurangkan tahun lahir dari tahun saat ini.

2. **Pembuatan Kelompok Usia:**
   - Usia kemudian dikelompokkan ke dalam rentang usia yang berbeda. Ini dilakukan untuk menyederhanakan analisis dan memungkinkan perbandingan antara kelompok usia yang berbeda.
   - Rentang usia yang dipilih yaitu:
    1. **Anak-anak dan Remaja:** Di bawah 17 tahun.
    2. **Dewasa Muda:** 18-34 tahun.
    3. **Dewasa:** 35-54 tahun.
    4. **Lansia:** 55 tahun ke atas.
        


```python
from datetime import datetime

# Menghitung usia berdasarkan 'Year_Birth'
# Asumsi data diambil pada tahun 2023
current_year = datetime.now().year
df['Age'] = current_year - df['Year_Birth']

# Mengelompokkan tahun lahir ke dalam kategori usia
new_bins = [0, 17, 34, 54, 100]
new_labels = ['Anak-anak/Remaja', 'Dewasa Muda', 'Dewasa', 'Lansia']
df['Age_Group'] = pd.cut(df['Age'], bins=new_bins, labels=new_labels, right=False, include_lowest=True)
```

Selanjutnya kita akan mengelompokkan jumlah pendapatan pelanggan menggunakan metode kuartil. Kuartil adalah nilai-nilai yang membagi data menjadi empat bagian yang sama besar.

2. **Pembagian Kelompok:**
   - **Kuartil Pertama (Q1):** 25% data terendah. Kelompok "Low" mencakup pendapatan dari nilai minimum hingga Q1.
   - **Kuartil Kedua (Median, Q2):** Memisahkan data menjadi dua bagian yang sama besar. Kelompok "Medium-Low" mencakup pendapatan dari Q1 hingga Q2.
   - **Kuartil Ketiga (Q3):** 75% data terendah. Kelompok "Medium-High" mencakup pendapatan dari Q2 hingga Q3.
   - **Kuartil Keempat:** Sisanya hingga nilai maksimum. Kelompok "High" mencakup pendapatan dari Q3 hingga nilai maksimum.

3. **Tujuan Pembagian:**
   - Tujuan dari pembagian ini adalah untuk memahami pola pengeluaran di berbagai tingkat pendapatan.
   - Ini membantu mengidentifikasi bagaimana pengeluaran produk berbeda di antara kelompok pendapatan yang berbeda, yang berguna untuk analisis pasar dan pengambilan keputusan pemasaran.

```python
# Mengelompokkan pendapatan ke dalam kuartil
Q1, Q2, Q3 = df['Income'].quantile([0.25, 0.5, 0.75])

# Membuat kelompok pendapatan berdasarkan kuartil
df['Income_Group'] = pd.cut(df['Income'], 
                            bins=[df['Income'].min()-1, Q1, Q2, Q3, df['Income'].max()], 
                            labels=['Low', 'Medium-Low', 'Medium-High', 'High'],
                            right=True)
```
